```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('census.csv')
df.head()
```

```python
big_states = df
big_states = big_states[["STNAME" , "CENSUS2010POP" ]]
big_states = big_states.sort_values(by=['CENSUS2010POP'], ascending=False)
first_fives = big_states[:5]
first_fives
```

```python
first_fives.set_index('STNAME', inplace=True)
plot = first_fives.plot.pie(y='CENSUS2010POP', autopct='%1.0f%%', figsize=(10, 10) , title="Censuses of Top Five Big States" )
plot
```

```python
df = df[["SUMLEV" ,"REGION" , "DIVISION" , "STATE" , "COUNTY" , "STNAME" , "CTYNAME" ,"CENSUS2010POP"]]
df
```

```python
%%timeit -n 5

for state in df['STNAME'].unique():
    avg = np.average(df[df['STNAME']==state]['CENSUS2010POP'])
    print('Counties in state ' + state + 
          ' have an average population of ' + str(avg))
```

```python
%%timeit -n 5

for group, frame in df.groupby('STNAME'):
    
    avg = np.average(frame['CENSUS2010POP'])
    print('Counties in state ' + group + 
          ' have an average population of ' + str(avg))
```
